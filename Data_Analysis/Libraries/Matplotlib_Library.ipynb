  - **Seaborn** üåä: Built on Matplotlib, beautiful defaults, great for statistical plots  
  - **Plotly** üåê: Interactive, dynamic, and web-based visualizations  

- **üåç Real-world Use Cases**  
  - üìà Analyzing trends and patterns in datasets  
  - üß™ Visualizing model performance and results

### Make both axes bold (thicker)

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random scores
scores = np.random.normal(loc=70, scale=10, size=100)
scores = np.clip(scores, 0, 100)

# Create figure and axes
fig, ax = plt.subplots()

# Plot histogram
ax.hist(scores, bins=10, color="lightgreen", edgecolor="black")

# Title and labels
ax.set_title("Exam Scores")
ax.set_xlabel("Score")
ax.set_ylabel("Number of Students")

# Remove grid
ax.grid(False)

# Make X-axis and Y-axis lines (spines) bold
ax.spines['bottom'].set_linewidth(3)
ax.spines['left'].set_linewidth(3)

plt.show()
```

### Make all four sides bold

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random scores
scores = np.random.normal(loc=70, scale=10, size=100)
scores = np.clip(scores, 0, 100)

# Create figure and axes
fig, ax = plt.subplots()

# Plot histogram
ax.hist(scores, bins=10, color="lightgreen", edgecolor="black")

# Title and labels
ax.set_title("Exam Scores")
ax.set_xlabel("Score")
ax.set_ylabel("Number of Students")

# Remove grid
ax.grid(False)

# Make all four sides bold
for spine in ax.spines.values():
    spine.set_linewidth(3)


plt.show()
```

```python
names = ['A', 'B', 'C']
values = [10, 20, 15]

bars = plt.bar(names, values)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,  # X position: center of the bar
        height, # Y position: top of the bar
        height, # Text to display: the height number
        ha='center', # Horizontal alignment: center
        va='bottom'  # Vertical alignment: bottom
    )

plt.show()
```

### StackPlot

```python
import matplotlib.pyplot as plt

years = [2018, 2019, 2020, 2021, 2022]
male = [40, 45, 50, 55, 60]
female = [35, 40, 45, 50, 55]

plt.stackplot(
    years,
    male,
    female,
    colors=['pink', 'cyan'],
    labels=['Male', 'Female']
)

plt.legend()
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("Stackplot Example")
plt.show()
```

### Fill Area

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 18, 25]

plt.plot(x, y)
plt.fill_between(x, y, color='lightblue', alpha=0.5)

plt.title("Fill Between Example")
plt.show()
```

### Rotate Text on X axis

```python
names = ['January', 'February', 'March', 'April']
values = [10, 20, 15, 25]

plt.bar(names, values)
plt.xticks(rotation=45)
plt.show()
```
