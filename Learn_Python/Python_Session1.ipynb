# Data Types

```python
print(type(10))    # int => integer
print(type(100))   # int => integer
print(type(-50))   # int => integer
print(type(23.34))  # float => Floating Point Number
print(type(-23.34))  # float => Floating Point Number
print(type("Hello Python"))  # str => String
print(type([1, 2, 3, 4, 5]))  # List => List
print(type((1, 2, 3, 4, 5)))  # Tuple => Tuple
print(type({1, 2, 3, 4, 5}))  # Set => Set
print(type({"name": "John", "age": 36}))  # Dict => Dictionary
print(type(True))  # bool => Boolean
print(type(False))  # bool => Boolean
```

# ‚úÖ Rules for Writing Variables in Python

(these are strict, you must follow them)

1. A variable name **must start** with a letter `(a-z, A-Z)` or an underscore `_`.

   - ‚úÖ `age`, `_value`  
   - ‚ùå `1age`, `#name`  

2. The rest of the name can have **letters, numbers, or underscores**.

   - ‚úÖ `age1`, `total_sum`, `car_speed`  
   - ‚ùå `car-speed`, `total$`  

3. Variable names are **case-sensitive**.

   - `Age`, `AGE`, and `age` are three different variables.  

4. You **cannot use reserved keywords** (like `if`, `while`, `class`, `for`) as variable names.  

---

# ‚ú® Best Practices (Recommended)

- Use **descriptive names** ‚Üí makes code readable.  
  - ‚úÖ `student_age`, `total_price`  
  - ‚ùå `x`, `abc` (unless used for math/short code)  

- Use **snake_case** (lowercase letters with underscores) for variables.  
  - ‚úÖ `first_name`  
  - ‚ùå `FirstName`, `firstName` (those styles are more common in other languages like Java or C#)  

- Keep names **short but meaningful**.
