# Project Build a Pyramid

````python
    #
   ###
  #####
 #######
#########
    #
# tall : 5
# 4 spaces : 1 hash
# 3 spaces : 3 hash
# 2 spaces : 5 hash
# 1 space  : 7 hash
# 0 spaces: 9 hash
````

````python
tree_height = int(input("Enter the height of the pyramid: "))
spaces = tree_height - 1
hash = 1
original_spaces = tree_height - 1

while tree_height != 0:
   for i in range(spaces):
    print(" ", end="")

   for i in range(hash):
    print("#", end="")

   print()
   spaces -= 1
   hash += 2
   tree_height -= 1

for i in range(original_spaces):
  print(" ", end="")
print("#")
````

A **set** in Python does not support list-specific methods such as:

- `append()`
- `insert()`

This is because a set is designed to store **unique elements**, and it does **not maintain order**, so inserting at a specific position does not make sense.

However, sets do have their **own methods** that you can use to modify and work with set elements.  
These include:

- `add()` â€” adds a single element  
- `update()` â€” adds multiple elements  
- `remove()` â€” removes an element (raises an error if not found)  
- `discard()` â€” removes an element (without error)  
- `pop()` â€” removes a **random** element  

### Mathematical operations:
- `union()`  
- `intersection()`  
- `difference()`  
- `symmetric_difference()`  

So while sets do not have list-like methods such as `append()` or `insert()`, they do provide their own built-in methods designed for working with **unordered unique collections**.


````python
set1 = {1, 2, 3, 4, 5}
print("Set 1 : ", set1)
set1.add(6)
print("Set1 after adding 6 : ",set1) # {1, 2, 3, 4, 5, 6}
set1.remove(3)
print("Set1 after removing 3 : ", set1) # will raise an error if the key is not found
````

````python
set1.discard(4) # doesn't raise an error if the element doesn't found
print("Discarding 4 : ", set1)
````
